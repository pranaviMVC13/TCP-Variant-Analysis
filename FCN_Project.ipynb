##### Experiment 1: Comparing through put of Tahoe, Reno and Cubic.
Bandwidths : 1, 3, 5, 7, 9.
Running the experiment for 3 times
    

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## cwnd of tahoe,reno and cubic at bandwidth 1MBps

dataset_Linux1 = open("fcn/Final_Exp1/cwnd_Tahoe_1Mbps_0","r")
dataset_Linux2 = open("fcn/Final_Exp1/cwnd_Reno_1Mbps_0","r")
dataset_Linux3 = open("fcn/Final_Exp1/cwnd_Linux_1Mbps_0","r")
time1 = []
value1 = []
time2 = []
value2 = []
time3 = []
value3 = []
for line in dataset_Linux1:
    columns = line.split()
    time1.append(columns[0])
    value1.append(float(columns[1]))
plt.plot(time1,value1,'-b',label= "Tahoe")

for line in dataset_Linux2:
    columns = line.split()
    time2.append(columns[0])
    value2.append(float(columns[1]))
plt.plot(time2,value2,'-g',label= "Reno")

for line in dataset_Linux3:
    columns = line.split()
    time3.append(columns[0])
    value3.append(float(columns[1]))
plt.plot(time3,value3,'-r',label= "Cubic")

plt.xlabel("time")
plt.ylabel("cwnd")
plt.legend(loc='upper left')
plt.autoscale(False)
plt.show()
```

cwnd analysis:

Tahoe upon a packet loss, sets the start state to 1MSS and thereshold to half of cwnd. This makes tahoe unscalable as we are moving the start to 1MSS everytime a loss occurs.  

Reno reduces the thresold to cwnd and start to half of cwnd.

Cubic's cwnd is calculated as a cubic growth function depending on the last congestion event. This is highly scalable especially for the flows with large RTTs.

```python
## cwnd of tahoe,reno and cubic at bandwidth 9MBps

dataset_Linux1 = open("fcn/Final_Exp1/cwnd_Tahoe_9Mbps_0","r")
dataset_Linux2 = open("fcn/Final_Exp1/cwnd_Reno_9Mbps_0_temp","r")
dataset_Linux3 = open("fcn/Final_Exp1/cwnd_Linux_9Mbps_0","r")
time1 = []
value1 = []
time2 = []
value2 = []
time3 = []
value3 = []
for line in dataset_Linux1:
    columns = line.split()
    time1.append(columns[0])
    value1.append(float(columns[1]))
plt.plot(time1,value1,'-b',label = "Tahoe")

for line in dataset_Linux2:
    columns = line.split()
    time2.append(columns[0])
    value2.append(float(columns[1]))
plt.plot(time2,value2,'-g', label = "Reno")

for line in dataset_Linux3:
    columns = line.split()
    time3.append(columns[0])
    value3.append(float(columns[1]))
plt.plot(time3,value3,'-r', label = "Cubic")

plt.xlim(0, 40)
plt.xlabel("time")
plt.ylabel("cwnd")
plt.legend(loc='upper right')
plt.autoscale(False)
plt.show()
```

###### experiment1 : throught vs bandwidth

```python
from collections import OrderedDict

dataset_Linux1 = open("fcn/Final_Exp1/Tahoe_0","r")
dataset_Linux2 = open("fcn/Final_Exp1/Reno_0","r")
dataset_Linux3 = open("fcn/Final_Exp1/Cubic_0","r")

value1 = {}
value2 = {}
value3 = {}
for line in dataset_Linux1:
    columns = line.split()
    time1 = int(columns[0][0])
    value1[time1] = float(columns[1])
value1 = sorted(value1.items(),key = lambda x : x[0])
time = [x[0] for x in value1]
value1 = [x[1] for x in value1]
plt.plot(time,value1,'-b',label = "Tahoe")

for line in dataset_Linux2:
    columns = line.split()
    time2 = int(columns[0][0])
    value2[time2] = float(columns[1])
value2 = sorted(value2.items(),key = lambda x : x[0])
value2 = [x[1] for x in value2]
plt.plot(time,value2,'-g', label = "Reno")

for line in dataset_Linux3:
    columns = line.split()
    time3 = int(columns[0][0])
    value3[time3] = float(columns[1])
print(value3)
value3 = sorted(value3.items(),key = lambda x : x[0])
value3 = [x[1] for x in value3]
plt.plot(time,value3,'-r', label = "Cubic")


plt.xlabel("CBR")
plt.ylabel("throughput")
plt.legend(loc='upper right')
plt.autoscale(False)
plt.show()
```

Tahoe waits till the third duplicate ACK is received(packet is dropped) for retransmission during packet loss and startts the cwnd from slow start phase changing the threshold to half of cwnd. As a result exhibits a bad throughput compared to others.

Reno overcomes the above problems by fast recovery. Reno doesnt start from slow start phase on packet loss instead retransmits once it recieves triple duplicate ACK and hence is better than Tahoe.

Cubic's throught put doesnt depend on packet loss 

###### experiment1 : link utilization vs CBR

```python
dataset_Linux1 = open("fcn/Final_Exp1/Tahoe_0","r")
dataset_Linux2 = open("fcn/Final_Exp1/Reno_0","r")
dataset_Linux3 = open("fcn/Final_Exp1/Cubic_0","r")

value1 = {}
value2 = {}
value3 = {}
for line in dataset_Linux1:
    columns = line.split()
    time1 = int(columns[0][0])
    value1[time1] = float(columns[2])
value1 = sorted(value1.items(),key = lambda x : x[0])
time = [x[0] for x in value1]
value1 = [x[1] for x in value1]
plt.plot(time,value1,'-b',label = "Tahoe")

for line in dataset_Linux2:
    columns = line.split()
    time2 = int(columns[0][0])
    value2[time2] = float(columns[2])
value2 = sorted(value2.items(),key = lambda x : x[0])
value2 = [x[1] for x in value2]
plt.plot(time,value2,'-g', label = "Reno")

for line in dataset_Linux3:
    columns = line.split()
    time3 = int(columns[0][0])
    value3[time3] = float(columns[2])
print(value3)
value3 = sorted(value3.items(),key = lambda x : x[0])
value3 = [x[1] for x in value3]
plt.plot(time,value3,'-r', label = "Cubic")


plt.xlabel("CBR")
plt.ylabel("Link Utilization")
plt.legend(loc='upper left')
plt.autoscale(False)
plt.show()
```

In tahoe, after packet loss, the cwnd starts from slow start and as a result, it will not be able to send as many number of packets as it can send using the full bandwidth.

Reno, instead of slow start, reduces cwnd by half and hence makes a better use of bandwidth compared to tahoe.

Cubic because of its high scalability, utilizes the bandwitdth better than tahoe and reno
